```python
import pandas as pd

data = pd.read_csv("SoftwareProject_DataSet.csv")
```

##Verificacion del contenido del DataSet

```python
data.head()
```

##Obtener el tamaño de la poblacion

---



```python
population_size = len(data)
print("The Poupulation size is over: ", population_size)
```

##Creacion de muestra aleatoria simple de 50 registros

---



```python
sample_size = 50
simple_sample = data.sample(n=sample_size)
```

##Revision de creacion de la muestra

---



```python
display(simple_sample)
```

##Demostrar el tamaño de la muestra

---



```python
sample_size = len(simple_sample)
print("The Sample size is over: ", sample_size)
```

##Descripcion estadistica de la muestra (Ejemplo)

---



```python
sample_description = simple_sample[['Desarrollador','Tiempo de desarrollo (horas)','Antigüedad del desarrollador (años)']].describe()
print("Descripcion estadistica de la muestra: ")
display(sample_description)
```

#Muestreo aleatorio sistematico

---



##Seleccion sistematica de una muestra

---



##Definir el intervalo K (tamaño poblacion / muestra deseada)

---



```python
sample_size = 80
k_interval = population_size // sample_size
```

##Seleccion de muestra sistematica

---



```python
systematic_sample = data.iloc[::k_interval]
```

##Verificacion de la muestra

---



```python
print("Muestreo obtenida a traves de muestreo aleatorio sistematic")
display(systematic_sample)
```

#Analisis de subpoblacion

---



##Dividir la poblacion por tipo de proyecto

---



```python
grouped_by_project_type = data.groupby('Tipo de proyecto')
```

##Calcular el promedio de lineas de codigo y bugs para cada subpoblacion

---



```python
group_analysis = grouped_by_project_type[['Líneas de código','Tiempo de desarrollo (horas)','Satisfacción del cliente']].mean()
print("Analisis de subpoblacion: ")
display(group_analysis)
```
